Looks like `x` and `y` have similar distributions...  So let's use a 1:1 aspect ratio for the plot. You can control this in the `Layout` object.

```python
layout = pgo.Layout(
    showlegend=False,
    hovermode='closest',
    xaxis1=pgo.XAxis(
        range=[-20.0, 120.0],
        domain=[0.0, 1.0],
        type='linear',
        showgrid=False,
        zeroline=False,
        showline=True,
        nticks=8,
        ticks='inside',
        tickfont=pgo.Font(
            size=10.0
        ),
        mirror='ticks',
        anchor='y1',
        side='bottom'
    ),
    yaxis1=pgo.YAxis(
        range=[-20.0, 120.0],
        domain=[0.0, 1.0],
        type='linear',
        showgrid=False,
        zeroline=False,
        showline=True,
        nticks=8,
        ticks='inside',
        tickfont=pgo.Font(
            size=10.0
        ),
        mirror='ticks',
        anchor='x1',
        side='left'
    )
)
```
