```python
import matplotlib.pyplot as plt
```

Let's represent some damped oscillation in time and space.

```python
import numpy as np
```

```python
# Time
t = np.linspace(0, 1, 400)
```

```python
# Position
gamma = 2  # damping coefficient
omega = 2 * np.pi / 0.25  # angular frequency
x = np.exp(-gamma * t) * np.cos(omega * t)
```

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots()
ax.scatter(t, x)
```

```python
# Signal
sig = 1000 * np.exp(1j * np.pi * x * 0.5)
```

```python
fig, ax = plt.subplots()
ax.scatter(t, x, s=np.real(sig))
```

```python
fig, ax = plt.subplots()
ax.scatter(t, x, s=np.real(sig), c=np.imag(sig))
```

```python
# Text annotation
ax.text(0.96, 0.96,
        'Color = imag(sig)\nSize = real(sig)',
        transform=ax.transAxes,
        horizontalalignment='right',
        verticalalignment='top',
        size=11)
```

```python
# Labels
ax.set_ylabel('position')
ax.set_xlabel('time')
```
